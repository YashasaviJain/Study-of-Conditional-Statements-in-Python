```python
#1a. Check wheather a number is poositive, negaative, or zero
num = int(input("Enter a Number:"))
if num < 0:
  print("Number is Negative")
elif num > 0:
  print("Number is Positive")
else:
  print("Zero")
```

```python
#2. Check wheather the given number is even or odd
num = int(input("Enter a Number:"))
if num % 2 == 0:
  print("Number is Even")
else:
  print("Number is Odd")
```

```python
#3. Find the largest of three number
a = int(input("Enter the value of a:"))
b = int(input("Enter the value of b:"))
c = int(input("Enter the value of c:"))
if a>=b and a>=c:
  print("Largest:",a)
elif b>=a and b>=c:
  print("Largest:",b)
else:
  print("Largest:",c)
```

```python
#4. Get input from user for a subject and
# Calculate grade using if-elif-else
marks = int(input("Enter Marks:"))
if marks >= 90:
  print("Grade A")
elif marks >= 75:
  print("Grade B")
elif marks >= 60:
  print("Grade C")
elif marks >= 40:
  print("Grade D")
else:
  print("Fail")
```

```python
#5. Check wheather a year is a leap year
year = int(input("Enter Year:"))
if (year%400==0) or (year%4==0 and year%100!=0):
  print("Leap Year")
else:
  print("Not a Leap Year")
```

```python
#6. Write a Python program to increment the date by 1year
from datetime import date

d = date(2024, 2, 29)
y = d.year + 1

if d.month == 2 and d.day == 29:
    new_date = date(y, 2, 28)
else:
    new_date = date(y, d.month, d.day)

print(d)
print(new_date)
```
